```python
aal = datasets.fetch_atlas_aal(version = 'SPM12')
aal_img = nib.load(aal.maps)
```

```python
sbc_img = nib.load(r'D:\course\fMRI_processing_May2024\CONN\groups\conn_g\results\firstlevel\SBC_01\BETA_Subject001_Condition001_Source001.nii')
data = sbc_img.get_fdata()
abs_data = np.abs(data)
data.shape
```

```python
value_thresh = 0.35 * np.max(abs_data)
value_thresh
```

```python
labels_indices = dict(zip(aal.labels, aal.indices))
```

```python
labels_indices
```

```python
related_roi = []
for label, index in labels_indices.items():
    region_mask = (aal_img.get_fdata() == float(index))
    region_img = new_img_like(aal_img, region_mask)
    masker = NiftiMasker(mask_img = region_img, target_affine = sbc_img.affine)
    value_of_each_voxel = np.abs(masker.fit_transform(sbc_img))
    count_voxel  = np.sum(value_of_each_voxel >= value_thresh)
    voxel_count_thr = round(0.20 * value_of_each_voxel.shape[1])
    if count_voxel >= voxel_count_thr:
        related_roi.append(label)
```
